```python
class Item:
    def __init__(self, name, quantity, price):
        self.name = name
        self.quantity = quantity
        self.price = price

    def __str__(self):
        return f"{self.name:<10} | {self.quantity:<8} | ${self.price:.2f}"

inventory = []

def add_item():
    name = input("Enter item name: ")
    quantity = int(input("Enter quantity: "))
    price = float(input("Enter price: "))
    inventory.append(Item(name, quantity, price))
    print("Item added successfully!\n")

def view_items():
    print("\n--- Inventory List ---")
    print("Name       | Quantity | Price")
    print("----------------------------")
    for item in inventory:
        print(item)
    print()

def update_item():
    name = input("Enter item name to update: ").lower()
    for item in inventory:
        if item.name.lower() == name:
            item.quantity = int(input("Enter new quantity: "))
            item.price = float(input("Enter new price: "))
            print("Item updated successfully!\n")
            return
    print("Item not found.\n")

def delete_item():
    name = input("Enter item name to delete: ").lower()
    for item in inventory:
        if item.name.lower() == name:
            inventory.remove(item)
            print(f"{item.name} deleted successfully.\n")
            return
    print("Item not found.\n")

def search_item():
    name = input("Enter item name to search: ").lower()
    for item in inventory:
        if item.name.lower() == name:
            print(f"Found: {item.name} | Quantity: {item.quantity}, Price: ${item.price:.2f}\n")
            return
    print("Item not found.\n")

def menu():
    while True:
        print("Inventory Menu:")
        print("1. Add item")
        print("2. View all items")
        print("3. Update item")
        print("4. Delete item")
        print("5. Search for item")
        print("6. Exit")
        choice = input("Enter your choice: ")

        if choice == "1":
            add_item()
        elif choice == "2":
            view_items()
        elif choice == "3":
            update_item()
        elif choice == "4":
            delete_item()
        elif choice == "5":
            search_item()
        elif choice == "6":
            print("Goodbye!")
            break
        else:
            print("Invalid choice. Please try again.\n")

menu()
```
