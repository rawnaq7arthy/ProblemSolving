```python
#Series = one column (1D).
#DataFrame = whole table (2D).

import pandas as pd

# Example of a Series
s = pd.Series([10, 20, 30, 40], index = ['Apple', 'Banana', 'Mango','peach'])
print("Series:")
print(s)

# Example of a DataFrame
df = pd.DataFrame({
    'Name': ['Ali', 'Asma', 'Amal'],
    'Age': [25, 30, 35],
    'City': ['Oman', 'Dubai', 'Qatar']
})
print("\nDataFrame:")
print(df)
```

```python
sales = pd.Series({
    'Apple': 10,
    'Banana': 20,
    'Mango': 30,
    'peach': 40})
print(sales)
```

```python
print(sales[2])
```

```python
print(sales['Apple'])
```

```python
print(sales[sales > 20])
```

```python
gro = pd.Series({'Milk':120, 'Bread':80, 'Eggs':200, 'Cheese':150, 'Butter':90 })
print(gro)
print("Total sales" ,gro.sum())
print("Product sold the most ",gro.idxmax())


maxgro = gro.max()
soldmost = gro[gro == maxgro]

print("Product sold the most:", soldmost)
```

```python
sales = pd.Series({
    'Shampoo': 150,
    'Soap': 90,
    'Toothpaste': 180,
    'Lotion': 120})


# Update "Soap" sales to 100
sales['Soap'] = 100

# Find all products where sales > 120
high_sales = sales[sales > 120]

print("Updated sales:\n", sales)
print("\nProducts with sales > 120:\n", high_sales)
```

```python
from datetime import datetime, date, time, timedelta
dt = datetime(2025, 4, 29, 14, 30, 0)
print(dt)
```

```python
#current date time
now = datetime.now()
print(now)
```

```python
#strptime --> output: default format of date and time
datestr = "29/04/2025"
dt = datetime.strptime(datestr, "%d/%m/%Y")
print(dt)
```

```python
#strftime --> output: given format of date and time
formatdate = dt.strftime("%d-%b-%Y %H:%M:%S")
print(formatdate)
```

```python
futuredate = dt + timedelta(days = 10)
print(futuredate)
```

```python
futuredate = dt + timedelta(weeks = 10)
print(futuredate)
```

```python
futuredate = dt + timedelta(hours=10)
print(futuredate)
```
