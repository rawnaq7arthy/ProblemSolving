```python
import random
import matplotlib.pyplot as plt

# Number of customers to simulate
num_customers = 5

# Define realistic time ranges for each stage (in minutes)
decision_time_range = (2, 20)          # Time customer takes to decide on the order
confirmation_time_range = (1, 15)      # Time the restaurant takes to confirm the order
preparation_time_range = (15, 60)      # Time to prepare the food
delivery_time_range = (5, 25)          # Time for delivery

# Simulate for each customer
total_times = []
for i in range(num_customers):
    decision_time = random.randint(*decision_time_range)
    confirmation_time = random.randint(*confirmation_time_range)
    preparation_time = random.randint(*preparation_time_range)
    delivery_time = random.randint(*delivery_time_range)

    total_time = decision_time + confirmation_time + preparation_time + delivery_time
    total_times.append(total_time)

    # Print individual customer timing breakdown
    print(f"Customer {i+1}:")
    print(f"  Decision Time: {decision_time} min")
    print(f"  Confirmation Time: {confirmation_time} min")
    print(f"  Preparation Time: {preparation_time} min")
    print(f"  Delivery Time: {delivery_time} min")
    print(f"  --> Total Time: {total_time} min\n")

# Calculate average total delivery time
average_time = sum(total_times) / num_customers
print(f"Average Total Delivery Time: {average_time:.2f} minutes")

# Plot the results
plt.figure(figsize=(10, 5))
plt.bar([f"Cust {i+1}" for i in range(num_customers)], total_times, color='olive')
plt.axhline(y=average_time, color='red', linestyle='--', label=f'Average = {average_time:.1f} min')
plt.title("Total Delivery Time per Customer")
plt.xlabel("Customer")
plt.ylabel("Total Time (minutes)")
plt.xticks(rotation=45)
plt.legend()
plt.tight_layout()
plt.show()
```
